```python
import gdown
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_id = '17hrrwD6gOLWnLOJRo3iSn6pyYtXbNfgW'
url = f'https://drive.google.com/uc?id={file_id}'

output = 'linkedin-jobs-africa.csv'
gdown.download(url, output, quiet=False)

df=pd.read_csv(output)
df.head()
```

```python
df.info()
df2=df.copy()
```

```python
df2['posted_date']=pd.to_datetime(df2['posted_date'])
```

```python
df2.info()
```

```python
df2['year_posted']=df2['posted_date'].dt.year
df2['month_posted']=df2['posted_date'].dt.month
```

```python
df2.info()
```

```python
df2.head()
```

```python
grouped=df2.groupby(['month_posted','onsite_remote']).size().reset_index(name='number_vacancies')
```

```python
pivot = grouped.pivot(index='month_posted', columns='onsite_remote', values='number_vacancies')
```

```python
pivot.plot(kind='bar')

plt.title('Vacancies per Month and type')
plt.ylabel('Number of vacancies')
plt.xlabel('Month')
plt.xticks(rotation=0)
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```
